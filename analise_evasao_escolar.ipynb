```python
# Fonte:
# INEP - Indicadores de Fluxo Escolar

import pandas as pd
import matplotlib.pyplot as plt

dados = {
    "Etapa de Ensino": [
        "Ensino Médio",
        "Ensino Médio (período posterior)",
        "Fundamental - Anos Finais",
        "Fundamental - Anos Iniciais"
    ],
    "Taxa de Evasão (%)": [
        11.1,
        9.1,
        4.3,
        1.5
    ]
}

df = pd.DataFrame(dados)

print(df)

plt.figure(figsize=(10,5))

plt.bar(
    df["Etapa de Ensino"],
    df["Taxa de Evasão (%)"]
)

plt.title("Taxas de Evasão Escolar no Brasil")
plt.ylabel("Taxa de Evasão (%)")
plt.xlabel("Etapa de Ensino")

plt.tight_layout()

plt.savefig("grafico_evasao.png")

plt.show()

print("Gráfico gerado com sucesso.")
```
